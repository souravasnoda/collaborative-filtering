```python
import pandas as pd
```

```python
ratings = pd.read_csv('ratings.csv')
movies = pd.read_csv('movies.csv')
ratings = pd.merge(movies,ratings).drop(['genres','timestamp'],axis = 1)
ratings.head()
```

```python
user_ratings = ratings.pivot_table(index=['userId'] , columns = ['title'],values = 'rating')
user_ratings.head()
```

```python
user_ratings.T
```

# REMOVING MOVIES HAVING LESS THAN 10 RATINGS

```python
user_ratings = user_ratings.dropna(thresh = 10 , axis = 1).fillna(0)#axis = 1 beacuse hame movies(column remove karni hain)
#if axis = 0 ralhenge to userid (yani rows remove hongi)
user_ratings
```

BUILDING SIMILARITY MATRIX

```python
item_similarity_df = user_ratings.corr(method = 'pearson')
item_similarity_df
```

```python
## let's makes Recomendation
def get_similar_movies(movie_name,user_rating):
    similar_score = item_similarity_df[movie_name]*user_rating
    similar_score = similar_score.sort_values(ascending = False)
    return similar_score


```

TESTING CODE

```python
action_lover = [("2 Fast 2 Furious (Fast and the Furious 2, The) (2003)",5) , ("12 Years a Slave (2013)",3), ("(500) Days of Summer (2009)",1)]
similar_movies  = pd.DataFrame()

for movie,rating in action_lover:
    similar_movies = similar_movies.append(get_similar_movies(movie,rating),ignore_index = True)
    
    
similar_movies.sum().sort_values(ascending = False)  
```
